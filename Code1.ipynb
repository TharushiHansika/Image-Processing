## Imcomplement - Inverts the image

```python
def imcomplement(I):
    return 255 - I
inverted_image = imcomplement(gray_image)
cv2_imshow(inverted_image)
```

## Flipud - Flips image along the x-axis

```python
def flipud(I):
    flipped = np.zeros_like(I)
    for i in range(I.shape[0]):
        flipped[i] = I[I.shape[0] - 1 - i]
    return flipped
flipped_image = flipud(gray_image)
cv2_imshow(flipped_image)
```

## Fliplr - Flips image along the y-axis

```python
def fliplr(I):
    flipped = np.zeros_like(I)
    for i in range(I.shape[1]):
        flipped[:, i] = I[:, I.shape[1] - 1 - i]
    return flipped
fliplr_image = fliplr(gray_image)
cv2_imshow(fliplr_image)
```

## Imresize - Resizes image with nearest-neighbor interpolation

```python
def imresize(I, new_size):
    x_scale = I.shape[1] / new_size[1]
    y_scale = I.shape[0] / new_size[0]
    resized = np.zeros(new_size, dtype=I.dtype)

    for i in range(new_size[0]):
        for j in range(new_size[1]):
            y = int(i * y_scale)
            x = int(j * x_scale)
            resized[i, j] = I[y, x]
    return resized
imresize_image = imresize(gray_image, (100, 200))
cv2_imshow(imresize_image)
```

## Rotation

```python
def rotate(image, angle):
    rows, cols = image.shape
    M = cv2.getRotationMatrix2D((cols / 2, rows / 2), angle, 1)
    rotated = cv2.warpAffine(image, M, (cols, rows))
    return rotated
rotated_image = rotate(gray_image, 45)
cv2_imshow(rotated_image)
```

## Shearing

```python
def shear(image, shear_factor):
    rows, cols = image.shape
    M = np.float32([[1, shear_factor, 0], [0, 1, 0]])
    sheared = cv2.warpAffine(image, M, (int(cols * (1 + abs(shear_factor))), rows))
    return sheared
sheared_image = shear(gray_image, 0.5)
cv2_imshow(sheared_image)
```

## Translation

```python
def translate(image, tx, ty):
    rows, cols = image.shape
    M = np.float32([[1, 0, tx], [0, 1, ty]])
    translated = cv2.warpAffine(image, M, (cols, rows))
    return translated
translated_image = translate(gray_image, 50, 30)
cv2_imshow(translated_image)
```

## Zoom

```python
def zoom_in(image, zoom_factor=2):
    h, w = image.shape
    center_x, center_y = w // 2, h // 2
    new_w, new_h = w // zoom_factor, h // zoom_factor
    x1, y1 = center_x - new_w // 2, center_y - new_h // 2
    x2, y2 = center_x + new_w // 2, center_y + new_h // 2
    cropped = image[y1:y2, x1:x2]
    zoomed_in = cv2.resize(cropped, (w, h), interpolation=cv2.INTER_LINEAR)
    return zoomed_in

zoomed_image = zoom_in(gray_image, zoom_factor=2)
cv2_imshow(zoomed_image)
```
